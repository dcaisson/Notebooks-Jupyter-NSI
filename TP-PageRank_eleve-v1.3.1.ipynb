# Comment sont classés les résultats d'une recherche ?

Le PageRank est l’algorithme d’analyse des liens hypertextes utilisé pour le classement des pages web par le moteur de recherche Google. Le PageRank n’est qu’un indicateur parmi d’autres dans l’algorithme qui permet de classer les pages du Web dans les résultats de recherche. Ce système a été inventé par Larry Page, cofondateur de Google. Ce mot est une marque déposée.

Le principe de base est d’attribuer à chaque page une valeur proportionnelle au nombre de fois que passerait par cette page un utilisateur parcourant le Web en cliquant aléatoirement sur un des liens apparaissant sur chaque page. À chaque instant, des « robots », appelés « spiders », parcourent la toile, de lien en lien, et mettent ainsi sans cesse à jour le PageRank des pages du web.
